## Load From PyTorch Hub

This example loads a pretrained **YOLOv5s** model and passes an image for inference. YOLOv5 accepts **URL**, **Filename**, **PIL**, **OpenCV**, **Numpy** and **PyTorch** inputs, and returns detections in **torch**, **pandas**, and **JSON** output formats. See the [YOLOv5 PyTorch Hub Tutorial](https://docs.ultralytics.com/yolov5/tutorials/pytorch_hub_model_loading/) for details.

```python
import torch

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Images
imgs = ['https://ultralytics.com/images/zidane.jpg']  # batch of images

# Inference
results = model(imgs)

# Results
results.print()
results.save()  # or .show()

results.xyxy[0]  # img1 predictions (tensor)
results.pandas().xyxy[0]  # img1 predictions (pandas)
#      xmin    ymin    xmax   ymax  confidence  class    name
# 0  749.50   43.50  1148.0  704.5    0.874023      0  person
# 1  433.50  433.50   517.5  714.5    0.687988     27     tie
# 2  114.75  195.75  1095.0  708.0    0.624512      0  person
# 3  986.00  304.00  1028.0  420.0    0.286865     27     tie
```

```python
# 1Ô∏è‚É£ Install dependencies
!pip install ultralytics matplotlib pillow mtcnn --quiet

# 2Ô∏è‚É£ Imports
from ultralytics import YOLO
from PIL import Image
from google.colab import files
import matplotlib.pyplot as plt
import os, glob
import numpy as np
from mtcnn import MTCNN
import torch
from cct_backbone import CCTBackbone  # <-- your custom backbone

# 3Ô∏è‚É£ COCO class names
COCO_INSTANCE_CATEGORY_NAMES = [
  '__background__', 'woman', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus',
  'train', 'truck', 'boat', 'plant', 'man', 'stop sign',
  'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep', 'cow',
  'elephant', 'bear', 'zebra', 'giraffe', 'backpack', 'umbrella', 'handbag',
  'tie', 'suitcase', 'frisbee', 'purse', 'snowboard', 'sports ball', 'kite',
  'baseball bat', 'baseball glove', 'skateboard', 'surfboard', 'tennis racket',
  'bottle', 'wine glass', 'cup', 'fork', 'knife', 'spoon', 'bowl', 'banana',
  'apple', 'sandwich', 'orange', 'broccoli', 'carrot', 'hot dog', 'pizza',
  'donut', 'cake', 'chair', 'couch', 'potted plant', 'bed', 'dining table',
  'toilet', 'tv', 'plant', 'mouse', 'remote', 'keyboard', 'cell phone',
  'microwave', 'oven', 'toaster', 'router', 'refrigerator', 'AP', 'clock',
  'vase', 'phone', 'teddy bear', 'hair drier', 'toothbrush'
]

# 4Ô∏è‚É£ Upload an image
uploaded = files.upload()
img_path = list(uploaded.keys())[0]
image = np.array(Image.open(img_path))

# 5Ô∏è‚É£ YOLO detection
model = YOLO("yolov8n.pt")
CONF_THRESH = 0.5
results = model.predict(source=img_path, conf=CONF_THRESH, iou=0.45, save=True, show=False)
res = results[0]

# Locate saved annotated image
output_dir = os.path.join("runs", "detect")
pred_dirs = sorted(glob.glob(os.path.join(output_dir, "predict*")), key=os.path.getmtime, reverse=True)
annotated_path = os.path.join(pred_dirs[0], img_path)

# Display annotated image
img = Image.open(annotated_path)
plt.figure(figsize=(12, 8))
plt.imshow(img)
plt.axis("off")
plt.show()

# Print detected objects
print(f"Detected objects (conf >= {CONF_THRESH}):")
for box in res.boxes:
    cls = int(box.cls)
    conf = float(box.conf)
    if conf >= CONF_THRESH:
        class_name = COCO_INSTANCE_CATEGORY_NAMES[cls] if cls < len(COCO_INSTANCE_CATEGORY_NAMES) else f"Class_{cls}"
        xyxy = [round(float(x), 2) for x in box.xyxy[0].tolist()]
        print(f"- {class_name}: {conf:.4f} | Box: {xyxy}")

# 6Ô∏è‚É£ Facial recognition with MTCNN
detector = MTCNN()
faces = detector.detect_faces(image)

for i, face in enumerate(faces):
    x, y, w, h = face['box']
    face_crop = image[y:y+h, x:x+w]
    face_pil = Image.fromarray(face_crop)
    plt.figure(figsize=(4,4))
    plt.imshow(face_pil)
    plt.axis("off")
    plt.title(f"Facial Recognition {i+1}")
    plt.show()

print(f"Total close-ups generated: {len(faces)}")

# 7Ô∏è‚É£ Pass image through CCTBackbone
backbone = CCTBackbone()
backbone.eval()  # inference mode

# Prepare tensor (1,3,H,W) normalized
img_tensor = torch.from_numpy(image).permute(2,0,1).unsqueeze(0).float() / 255.0

with torch.no_grad():
    features = backbone(img_tensor)

# Visualize backbone feature maps
for i, feat in enumerate(features):
    # Take mean over channels to make a single grayscale map
    feat_map = feat.squeeze(0).mean(0).numpy()
    plt.figure(figsize=(6,6))
    plt.imshow(feat_map, cmap='viridis')
    plt.axis("off")
    plt.title(f"CCTBackbone Feature Map Stage {i+1} ({feat.shape[1]}x{feat.shape[2]})")
    plt.show()
```
